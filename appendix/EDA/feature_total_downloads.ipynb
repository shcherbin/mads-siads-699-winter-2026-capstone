## Exploratory Data Analysis (EDA) - `total_downloads` Feature

```python
import numpy as np
import polars as pl
import altair as alt
from settings import load_settings
```

```python
alt.data_transformers.enable("vegafusion")
```

```python
SETTINGS = load_settings()
PYPI_TOTAL_DOWNLOADS = SETTINGS.feature_total_downloads_path
```

```python
pypi_td_df = pl.scan_parquet(PYPI_TOTAL_DOWNLOADS)
pypi_td_df.collect_schema()
```

```python
pypi_td_df.head().collect()
```

```python
pypi_td_df.select(pl.col("package_name")).unique().count().collect()
```

```python
pypi_td_df.select("total_downloads").describe()
```

```python
pdf = pypi_td_df.collect().to_pandas()
pdf["log_downloads"] = np.log10(pdf["total_downloads"] + 1)
```

```python
chart = (
    alt.Chart(pdf)
    .mark_boxplot()
    #.mark_boxplot(extent="min-max")
    .encode(
        x=alt.X(
            "total_downloads:Q",
            title="Total downloads",
            scale=alt.Scale(type="log")
        )
    )
    .properties(
        title="Tukey boxplot of total_downloads",
        width=800,
        height=100
    )
)

chart
```

```python
chart = (
    alt.Chart(pdf)
    .mark_bar()
    .encode(
        alt.X("log_downloads:Q", bin=alt.Bin(maxbins=50)),
        y="count()",
    )
    .properties(
        title="Histogram of total_downloads (log scale)",
        width=800,
        height=400
    )
)

chart
```
